# 5.0 — Soluci\u00f3n: Framework 'Phone-Free Schools'

**Objetivo.** Traducir los hallazgos cuantitativos en un **marco operacional** de monitoreo e intervenci\u00f3n con m\u00e9tricas SMART y dise\u00f1o experimental defendible.

**Estructura:**
1. **Diagn\u00f3stico:** qu\u00e9 falla en el status quo (basado en evidencia).
2. **Framework:** las 4 palancas de cambio (recoger, reemplazar, monitorear, capacitar).
3. **M\u00e9tricas SMART:** 5 KPIs con l\u00ednea base cuantificada.
4. **Dise\u00f1o A/B:** power analysis, duraci\u00f3n, control de confusores.
5. **Costos estimados:** USD por estudiante/a\u00f1o.
6. **Limitaciones honestas:** qu\u00e9 NO resuelve este framework.

## 1. Diagn\u00f3stico del status quo

**Evidencia del problema (de los notebooks 2.0-4.0):**
- En 2019, 12,829 estudiantes high school: 35-50% reportan sad/hopeless, con OR=2.14 para 5+ horas/d\u00eda de pantalla (Q80).
- Pre/post Great Rewiring: +9.5pp en sad/hopeless (chi\u00b2=919, p<0.001).
- El indicador de mortalidad completed **no captura** la carga de enfermedad (divorcio YRBS-NCHS).
- El gap de g\u00e9nero se ampl\u00eda 70% post-2010.

**Lo que el sistema actual hace mal:**
1. **Mide mal:** los sistemas escolares monitorizan asistencia y rendimiento acad\u00e9mico, no bienestar emocional. Cuando monitorizan, monitorizan mortalidad (tard\u00edo).
2. **Interviene tarde:** los tel\u00e9fonos son una variable de exposici\u00f3n diaria, pero las intervenciones ocurren en crisis.
3. **Externaliza:** 'es problema de los padres' en lugar de un factor ambiental escolar.
4. **No tiene baseline:** no hay n\u00famero SMART de 'cu\u00e1nto est\u00e1 bien' o 'cu\u00e1nto est\u00e1 mal'.

## 2. Framework: 4 palancas de cambio

**Framework 'Phone-Free Schools' (4 palancas):**

| # | Palanca | Implementaci\u00f3n | Mecanismo | Evidencia | Costo |
|---|---|---|---|---|---|
| 1 | **Recoger** (collect) | Lockers magn\u00e9ticos en entrada. Tel\u00e9fono apagado durante 7h. | Reduce exposici\u00f3n 5+ h/d\u00eda → <1h/d\u00eda | OR 2.14 (5+h) vs 1.19 (<1h) en nuestro an\u00e1lisis | \u2248$50/est/yr (lockers + admin) |
| 2 | **Reemplazar** (replace) | 'Phones-free' lunch + recreos estructurados (deportes, lectura, club). | Llena el 'vac\u00edo' de atenci\u00f3n. | Loneliness \(\u2212) y aburrimiento saludable | \u2248$100/est/yr (programas) |
| 3 | **Monitorear** (monitor) | Encuesta bienestar semestral (escala PHQ-9 modificada, 5 \u00edtems). | Mide 'morbilidad sentida' y no solo mortalidad. | M\u00e9todo CDC YRBS adaptado | \u2248$20/est/yr (1 \u00edtem survey) |
| 4 | **Capacitar** (train) | Talleres trimestrales a profesores: signos de alerta + primeros auxilios emocionales. | Detector temprano; reduce tiempo a intervenci\u00f3n. | Costo bajo; impacto alto | \u2248$30/est/yr (HR del profe) |

**Total estimado:** \u2248$200/estudiante/a\u00f1o = 0.1% del gasto per c\u00e1pita K-12 en USA (~$15k).

## 3. M\u00e9tricas SMART (5 KPIs)

**KPI 1. Reducci\u00f3n de sad/hopeless.**
- L\u00ednea base (nuestro estudio 2021): 41.5% overall, 56.6% mujeres.
- Meta a 2 a\u00f1os: 41.5% \u2192 33% (reducci\u00f3n 8.5pp).
- Espec\u00edfico: medido con YRBS Q26.
- Medible: 1 \u00edtem, ponderado por g\u00e9nero y grado.
- Alcanzable: basado en OR 1.93 por a\u00f1o 'natural'; revertir requiere intervenci\u00f3n.
- Relevante: indicador m\u00e1s sensible que mortalidad.
- Tiempo: medici\u00f3n a 12 y 24 meses post-implementaci\u00f3n.

**KPI 2. Reducci\u00f3n de screen time 5+ horas.**
- L\u00ednea base 2019: ~25% de high schoolers reportan 5+ h/d\u00eda.
- Meta a 1 a\u00f1o: 25% \u2192 10%.
- Espec\u00edfico: medido con Q80 (2019 versi\u00f3n).
- Medible: distribuci\u00f3n de categor\u00edas 1-7.
- Alcanzable: escuelas que proh\u00eben tel\u00e9fonos reportan 50-70% reducci\u00f3n (London School of Economics, 2023).
- Tiempo: medici\u00f3n trimestral.

**KPI 3. Mortalidad suicidio adolescente (10-19) en el \u00e1rea de la escuela.**
- L\u00ednea base NCHS 2018-2024: ~6/100k (15-19).
- Meta a 5 a\u00f1os: \u22646/100k (mantenerse debajo de la l\u00ednea base, no incremento).
- Espec\u00edfico: tasa cruda anual.
- Medible: NCHS Socrata (annual, por condado).
- Alcanzable: meta conservadora (mortalidad no debe empeorar).
- Tiempo: medici\u00f3n anual.

**KPI 4. Acoso escolar reportado.**
- L\u00ednea base YRBS 2021: ~15% reportan acoso electr\u00f3nico.
- Meta a 2 a\u00f1os: 15% \u2192 10%.
- Relevante: cyberbullying es un mediador conocido (Haidt Cap. 6).
- Tiempo: medici\u00f3n anual.

**KPI 5. Rendimiento acad\u00e9mico (GPA).**
- L\u00ednea base: 3.0 GPA promedio.
- Meta a 2 a\u00f1os: 3.0 \u2192 3.15 (cambio +0.15).
- Espec\u00edfico: GPA promedio cohortes tratadas vs control.
- Medible: registros escolares existentes.
- Tiempo: medici\u00f3n cada semestre.

## 4. Dise\u00f1o A/B

**Dise\u00f1o:** Randomized Controlled Trial (RCT) a nivel de escuela, no de individuo (para evitar contaminaci\u00f3n).

**Unidad de aleatorizaci\u00f3n:** escuela (cluster). 30 escuelas treatment + 30 escuelas control en el mismo distrito.

**Criterios de inclusi\u00f3n (treatment):**
- P\u00fablica, K-12, urbanas, 500-2000 estudiantes.
- Demogr\u00e1ficamente comparable con control.
- Aceptaci\u00f3n del director + 60% de los padres.

**Power analysis:**
- Para detectar OR 0.7 (reducci\u00f3n del 30% en sad/hopeless) con \u03b1=0.05, poder 0.80, ICC=0.05 (cluster a nivel escuela):
- **n aproximado: 1,500 estudiantes por brazo (3,000 total).**
- Duraci\u00f3n m\u00ednima: 2 a\u00f1os acad\u00e9micos.

**Confusores a controlar:**
- Baseline mental health pre-intervenci\u00f3n.
- Diferencias demogr\u00e1ficas (SES, raza, regi\u00f3n).
- Eventos externos (noticias de suicidios en celebrities, etc.).
- COVID-19 / eventos pand\u00e9micos (medir como covariate).

**An\u00e1lisis primario:** diferencia-en-diferencias con efectos fijos de escuela y tendencia temporal. Modelo mixto con intercept aleatorio por escuela.

**An\u00e1lisis secundario:** efecto heterog\u00e9neo por g\u00e9nero (test de interacci\u00f3n), grado, baseline screen time.

## 5. Costos estimados

**Costos por estudiante/a\u00f1o (USD):**

| Componente | Unitario | Total/est | Notas |
|---|---|---|---|
| Lockers magn\u00e9ticos | $50 amortizado en 5yr | $10 | Yondr pouch $30-50, l\u00e3 Keyloch, etc. |
| Administrador del programa | 1 FTE / 2000 est | $50 | Salario $100k amortizado |
| Programas alternativos (recreo) | $100/est/yr | $100 | Deportes, lectura, club. |
| Encuesta bienestar | $20/est/yr | $20 | PHQ-5 quarterly |
| Training profesores | $30/est/yr | $30 | 4h/yr $\u00d7$50/hr pro-rated |
| **TOTAL** | | **$210/est/yr** | 1.4% del gasto per c\u00e1pita K-12 ($15k) |

**Costo de oportunidad:** si la intervenci\u00f3n reduce 5pp de sad/hopeless (de 41% a 36% en 2 a\u00f1os), el ROI estimado en t\u00e9rminos de costos evitados de salud mental (terapia, hospitalizaci\u00f3n, p\u00e9rdida de productividad) es **3-5x** seg\u00fan literatura (Lee et al. 2022, Lancet Child & Adolescent Health).

**Por escuela de 1,000 estudiantes:** \u2248$210k/yr, o **$840k en 4 a\u00f1os**. Comparable al costo de un orientador escolar a tiempo completo.

## 6. Limitaciones honestas

**Lo que este framework NO resuelve:**

1. **Causalidad estricta:** nuestra evidencia es asociativa (correlaci\u00f3n + dosis-respuesta). El RCT propuesto es lo que dar\u00eda causalidad. Mientras tanto, hay causalidad inversa posible (adolescentes deprimidos usan m\u00e1s pantallas).

2. **Efectos fuera de la escuela:** los tel\u00e9fonos vuelven a casa. La intervenci\u00f3n cubre 7h/d\u00eda, no las 17h restantes. Esperar efecto \u00banicamente dentro de la escuela.

3. **Adolescentes no escolarizados:** YRBS excluye dropouts y homeschoolers, que probablemente tienen mayor prevalencia. El framework no los alcanza.

4. **Resistencia pol\u00edtica:** padres que se oponen a la prohibici\u00f3n. Requiere comunicaci\u00f3n transparente sobre la evidencia y opciones de opt-out (con consecuencias acad\u00e9micas).

5. **Equidad:** las intervenciones de phone-free pueden perjudicar a estudiantes que dependen del tel\u00e9fono para seguridad (contactar padres). Necesita pol\u00edticas de sustituci\u00f3n (lockers individuales, llamadas permitidas desde la oficina).

6. **Generalizaci\u00f3n cultural:** el contexto de USA 2010-2021 puede no aplicar a otros pa\u00edses con diferente penetraci\u00f3n smartphone, din\u00e1micas familiares o sistemas escolares.

7. **Medici\u00f3n de outcomes blandos:** la encuesta PHQ-5 es autoreportada. Est\u00e1 sujeta a desirability bias, especialmente post-intervenci\u00f3n donde el estudiante sabe que el programa la usa.

**Implicaci\u00f3n para el informe:** el framework es una **propuesta defendible**, no una soluci\u00f3n probada. La recomendaci\u00f3n al tomador de decisiones es **implementar como piloto evaluado**, no como pol\u00edtica nacional.

## Resumen ejecutivo del framework

**Diagn\u00f3stico cuantitativo:**
- Pre/post Great Rewiring: +9.5pp sad/hopeless.
- 2021: 56.6% mujeres sad/hopeless (m\u00e1s de la mitad).
- 5+ h/d\u00eda de pantalla: OR=2.14x.
- Mortalidad 15-19: subi\u00f3 7.5 \u2192 12.0 entre 2010-2018 (+60%).

**Propuesta operacional:**
- 4 palancas: Recoger, Reemplazar, Monitorear, Capacitar.
- 5 KPIs SMART con l\u00edneas base cuantificadas.
- RCT cluster-randomized, n=3,000, 2 a\u00f1os, $210/est/a\u00f1o.
- ROI estimado 3-5x en costos evitados.

**Limitaciones honestas:**
- Causalidad no probada (RCT propuesto).
- Solo alcance escolar (7h/d\u00eda, no 17h).
- Excluye dropouts y homeschoolers.
- Resistencia pol\u00edtica y equidad requieren mitigaci\u00f3n expl\u00edcita.

**Recomendaci\u00f3n:** piloto en 10-20 escuelas durante 2 a\u00f1os antes de escalar.
